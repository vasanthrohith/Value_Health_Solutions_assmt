```python
# pip install wikipedia
# pip install pyjokes

from datetime import datetime
import wikipedia
import pyjokes
from time import time, ctime


class Chatbot():
    global chatbot_name, exit_cmd,t,wikipedia_cmd,hello_cmd,jokes_cmd,asktime_cmd
    t = datetime.now().strftime('%H:%M %p')
    # commands
    hello_cmd=["hi","hai","hello"]
    chatbot_name = ["what is your name","tell me your name","what's your name","your name?"]
    exit_cmd=["quit","exit","close","bye"]
    wikipedia_cmd=["give me infos","explain me","search","wikipedia","wiki"]
    jokes_cmd = ['joke', 'tell me a joke', 'say me a joke', "it's boring", 'tell me joke', 'say me joke',
                 'tell a joke', 'say a joke']
    asktime_cmd=["what is the time now","time please","time","tell me the time"]

    def greetings(self):
        # greetings for the day
        t = datetime.now().strftime('%H')
        # print(t)
        t = int(t)
        # print(t)

        if t >= 4 and t < 12:
            print('ChatBot: Hi, Good morning.')
        elif t >= 12 and t < 16:
            print('ChatBot: Hi, Good afternoon.')
        elif t >= 16 and t < 19:
            print('ChatBot: Hi, Good evening.')
        else:
            print('ChatBot: Hi, Good night.')
        print("How can i help you today?")

    def history(self,message):
        message_history = []
        if message=="history--":
            print(message_history)
        else:
            message_history.append(message)
        # print(message_history)

    def chat(self):
        def asktime():
            datetime = ctime(time()).split(' ')
            # print(datetime)
            # print("Date: " + str(datetime[4]) + "-" + str(datetime[1]) + "-" + str(datetime[2]))

            print("Date: "+str(datetime[3]) + "-" + str(datetime[1])+ "-" +str(datetime[5]))
            print("Time: "+str(datetime[4]))
            t = str(datetime[3]) + "-" + str(datetime[1])+ "-" +str(datetime[5]) +"  " +str(datetime[4])
            return t


        def jokes_fn():  # to say jokes
            joke = pyjokes.get_joke()
            print(joke)
            return joke


        def wiki(input_):
            try:
                search=input("search: ")
                info = wikipedia.summary(search, 1)
                print(info)
            except:
                info="None"
            return info

        def history(message):
            message_history = []
            # if message == "history--":
            #     print(message_history)
            # else:
            message_history.append(message)
            # print(message_history)
            file=open("C:\\Users\\vasanth rohith\\OneDrive - Kau Yan College\\Desktop\\chatbot_history\\history.txt",'a')
            file.write(f"\n You: {message[0]}  {message[2]} \n")

            file.write(f"Chatbot: {message[1]}  {message[2]}")

        # print(chatbot_name)
        # history=Chatbot()
        while True:
            user_input = input("You: ")
            input_="".join([i.lower() for i in user_input])
            # print(type(input_))

            if input_ in chatbot_name:
                reply=["My name is ChatBot"]
                r=reply[0]
                print('ChatBot: ',r)
                history([user_input,r,f"time :{t}"])
            elif input_ in wikipedia_cmd:
                user =user_input
                r1=wiki(input_)
                history([user, r1, f"time :{t}"])
            elif input_ == "history--":
                history("history--")
            elif input_ in hello_cmd:
                print("Hai!")
                history([input_,"Hai!", f"time :{t}"])
            elif input_ in jokes_cmd:
                user_0 = user_input
                jokes=jokes_fn()
                history([user_0, jokes, f"time :{t}"])
            elif input_ in asktime_cmd:
                user_1=input_
                time_1=asktime()
                history([user_1, time_1, f"time :{t}"])

            elif input_ in exit_cmd:
                break


chatbot=Chatbot()
chatbot.greetings()
chatbot.chat()

```
